```python
import json
# Query all columns
query = """
SELECT TABLE_NAME, COLUMN_NAME, DATA_TYPE, COLUMN_KEY
FROM INFORMATION_SCHEMA.COLUMNS
WHERE TABLE_SCHEMA = 'nw_realite'
ORDER BY TABLE_NAME, ORDINAL_POSITION;
"""
df_all_columns = pd.read_sql(query, conn)

# Convert to dictionary in the format {"table": [columns]}
json_dict = {}
for table, group in df_all_columns.groupby("TABLE_NAME"):
    json_dict[table] = group["COLUMN_NAME"].tolist()

# Convert to JSON string (pretty printed)
json_output = json.dumps(json_dict, indent=2)
print(json_output)
```
